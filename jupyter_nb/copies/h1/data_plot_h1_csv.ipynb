## SANCTIONED INDIVIDUALS AND GENDER INFORMATION

Only individuals labelled as 'sanctioned' were considered from the statements.csv file. These individuals can be divided into two categories: sanctioned individuals and counter-sanctioned individuals. The latter differ from the former in that they are individuals who are sanctioned by non-democratic countries, which often include pro-democracy activists, journalists and human rights proponents in their own sanctions designations. In this context, however, both categories are considered together because, regardless of the sender, these individuals appear on official sanctions lists. **The total number of sanctioned individuals is 40900, of whom 9694 are counter-sanctioned.**

**The total of sanctioned individuals with gender information is 12046 (29.45%). The total without gender information is 28854 (70.55%). Hypothesis 1 is confirmed.** 
Opensanctions classifies gender into four categories: male, female, other and null. Figures 1 and Table 1 show the gender distribution among sanctioned individuals. The 'mixed' category is used for cases presenting more than one gender value. Often, the double value is the result of errors in the source data, but it is also possible that both values are attributed to the individual. Therefore, both values of these exceptional cases are taken into consideration in the total counting.
