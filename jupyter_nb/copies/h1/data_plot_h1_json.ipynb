## SANCTIONED INDIVIDUALS AND GENDER INFORMATION

Only individuals labelled as 'sanctioned' were considered from the datasets. These individuals can be divided into two categories: sanctioned individuals and counter-sanctioned individuals. The latter differ from the former in that they are individuals who are sanctioned by non-democratic countries, which often include pro-democracy activists, journalists and human rights proponents in their own sanctions designations. The total number of sanctioned individuals varies considerably from year to year primarily because more sanctions lists were added to the dataset over time (see Table 1 and Figure 1).

```python
from opensanctions.h1.h1_json import json_total_sanctioned_individual

json_total_sanctioned_individual()
```

The total number of sanctioned individuals with gender information also varies from year to year, most likely because sanctions lists were added later. However, we can see that, compared to the total number of individuals sanctioned each year, the distribution of individuals for whom gender information is available has decreased over time, and has never constituted the majority of cases. (see Table 2 and Figure 2). **Hypothesis 1 is confirmed.**
