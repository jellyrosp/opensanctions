Hypothesis 1: *senders do not consider gender a relevant category in sanctions listing data.*

The JSON files are a combination of [*Consolidated sanctions datasets*](https://www.opensanctions.org/datasets/sanctions/), representing the broadest set of sanctioned individuals designated globally. These datasets cover the period from 2021 to 2026 and were used to produce the following graphs and tables. These show relevant information extracted from the datasets that led to the confirmation of Hypothesis 1.

## SANCTIONED INDIVIDUALS AND GENDER INFORMATION

Only individuals labelled as 'sanctioned' were considered from the datasets. These individuals can be divided into two categories: sanctioned individuals and counter-sanctioned individuals. The latter differ from the former in that they are individuals who are sanctioned by non-democratic countries, which often include pro-democracy activists, journalists and human rights proponents in their own sanctions designations. In this context, however, both categories are considered together because, regardless of the sender, these individuals appear on official sanctions lists. Although the total number of sanctioned individuals varies considerably from year to year, this is primarily because more sanctions lists were added to the dataset over time (see Figure 1, Table 1).

**Figure 1, Table 1**

The total number of sanctioned individuals with gender information also varies from year to year because sanctions lists were added later. However, we can see that, compared to the total number of individuals sanctioned each year, the proportion of individuals for whom gender information is available has decreased over time, and has never constituted the majority of cases. (see Figure 2, table 2). **Hypothesis 1 is confirmed.**

Opensanctions classifies gender into four categories: male, female, other and null (see [here](https://github.com/opensanctions/opensanctions/blob/main/datasets/_externals/wikidata.yml#L125-L170) from row 250). Figures 3 and Table 3 show the gender distribution among sanctioned individuals for each year. The 'mixed' category is used for cases presenting more than one gender value. Often, the double value is the result of errors in the source data, but it is also possible that both values are attributed to the individual. Therefore, both values of these exceptional cases are taken into consideration in the total counting. Unlike the statement.csv dataset, it was not possible to discern the exact source of the gender information in the JSON files. However, this does not impact the confirmation of hypothesis 1, as the number of individuals with gender information remains significantly low.
