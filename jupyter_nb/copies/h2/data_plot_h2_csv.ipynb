# Hypothesis 2: CSV processing outcomes

Hypothesis 2: *women are frequently targeted due to close associations with other sanctioned entities.*

[Topics](https://www.opensanctions.org/docs/topics/) are labels used to classify individuals and indicate why they are included in a dataset. Table 1 and Figure 1 present an overview of the gender distribution across topics among sanctioned individuals. Topics are assigned a specific code. See [here](https://followthemoney.tech/explorer/types/topic/) for a list of the codes and their respective topics. 

Opensanctions assigns the **'role.rca'** topic to individuals who are primarily sanctioned due to their **close association** with politically exposed persons (topic code: **'role.pep'**), who are individuals who currently or previously held a position of political influence. The **'role.rca'** topic specifically captures the relationship-based rationale behind the designation of relatives or close associates.

```python
from opensanctions.h2.h2_csv import csv_sanctioned_gender_topics_frequency

csv_sanctioned_gender_topics_frequency()
```

To assess whether a specific topic is more commonly associated with one gender than the other, we measure its frequency within each gender group separately, ensuring that the analysis accounts for differences in group size. Table 2 and Figure 2 show the relative frequency of topics for the gender category.

**Analysis:**

- Males: 10407 individuals, with a total of 12046 entries
- Females: 1639 individuals, with a total of 12046 entries

*Topics with highest relative frequency:*

- Males: the most frequent topics are debarment* (60.32%), and politically-exposed person (role.pep) (36.72%)
- Females: the most frequent topics are debarment (47.90%), and politically-exposed person (role.pep) (42.02%)

*The individual is connected to companies that have been excluded from public procurement, often because they previously committed fraud in their execution of a government contract.

*Topics with the largest gender gaps:*

- Export control: males (29.70%) vs. females (15.17%)  +14.53% more frequent for males
- Debarment: males (60.32%) vs. females (47.90%)  +12.42% more frequent for males
- Military (mil): males (5.48%) vs. females (0.65%)  +4.83% more frequent for males

*Topics with higher relative frequency for females:*

- Close associate (role.rca): females (8.35%) vs. males (2.35%)  +6.00% more frequent for females
- Person of interest (poi)*: females (38.17%) vs. males (27.60%)  +10.57% more frequent for females

*An individual considered a public interest case for closer scrutiny, but who do not meet common definitions of a politically-exposed person and who are not necessarily sanctioned.

*Other differences:*

- Judge (role.judge): present only in females (1.52%)
- Oligarch (role.oligarch): rare for both, but slightly more frequent for males (1.15%) than females (0.15%)
- War crimes (crime.war): very rare for both, but slightly more frequent for males (0.85%) than females (0.07%)

**These findings indicate that women are more frequently sanctioned due to their close association with a sanctioned individual compared to men. This result confirms Hypothesis 2.**
