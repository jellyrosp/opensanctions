# Hypothesis 2: CSV processing outcomes

Hypothesis 2: *women are frequently targeted due to close associations with other sanctioned entities.*

[Topics](https://www.opensanctions.org/docs/topics/) are labels used to classify individuals and indicate why they are included in a dataset. Although topics help to assess the type of risk posed by an individual, they are not intended to reflect the specific reasons for the designation. However, the associated risk often aligns with the reasons, providing preliminary insights into why the person was sanctioned. Figure 1 and Table 1 provide an overview of the frequency of gender-related topics. Topics are assigned a specific code. See [here](https://followthemoney.tech/explorer/types/topic/) for a list of the codes and their respective topics. 

Opensanctions assigns the **'role.rca'** topic to individuals who are primarily sanctioned due to their **close association** with politically exposed persons (topic code: **'role.pep'**), who are individuals who currently or previously held a position of political influence. The **'role.rca'** topic specifically captures the relationship-based rationale behind the designation of relatives or close associates.

**Figure 1, Table 1**

To assess whether a specific topic is more commonly associated with one gender than the other, we measure its frequency within each gender group separately, ensuring that the analysis accounts for differences in group size. Figure 2 and Table 2 show the relative frequency of topics for the gender 'male'. Figure 3 and Table 3 show the relative frequency of topics for the gender 'female'.

**Analysis:**

- Males: 10407 individuals, with a total of 12046 entries
- Females: 1639 individuals, with a total of 12046 entries

*Topics with highest relative frequency:*

- Males: the most frequent topics are debarment (60.32%), and politician (role.pep) (36.72%)
- Females: the most frequent topics are debarment (47.90%), and politician (role.pep) (42.02%)

*Topics with the largest gender gaps:*

- Export control: males (29.70%) vs. females (15.17%)  +14.53% more frequent for males
- Debarment: males (60.32%) vs. females (47.90%)  +12.42% more frequent for males
- Military (mil): males (5.48%) vs. females (0.65%)  +4.83% more frequent for males

*Topics with higher relative frequency for females:*

- Close associate (role.rca): females (8.35%) vs. males (2.35%)  +6.00% more frequent for females
- Person of interest (poi) (public interest case for closer scrutiny, but who do not meet common definitions of a politically-exposed person and who are not necessarily sanctioned): females (38.17%) vs. males (27.60%)  +10.57% more frequent for females

*Other differences:*

- Judge (role.judge): present only in females (1.52%)
- Oligarch (role.oligarch): rare for both, but slightly more frequent for males (1.15%) than females (0.15%)
- War crimes (crime.war): very rare for both, but slightly more frequent for males (0.85%) than females (0.07%)

**These preliminary findings indicate that women are more frequently sanctioned due to their close association with a sanctioned individual compared to men. This result provides partial support for Hypothesis 2.**
