# Hypothesis 2: JSON processing outcomes

Compared to the CSV file, the JSON files provide significantly less information. For this reason, only nationality data were extracted, as the data for both religion and ethnicity were reported for fewer than 10 individuals. The same limitation applies to the topics data.

The data on nationality is not directly tied to Hypothesis 2, it was examined to identify potential trends. Due to the limited sample size of sanctioned individuals per category, the findings are not statistically robust enough to draw generalizable conclusions. Nevertheless, the absence of gender-disaggregated information highlights a significant methodological limitation in sanctions research. Although the importance of incorporating gender-sensitive analysis is widely acknowledged, the lack of comprehensive and systematically collected gender data substantially constrains the analytical scope and undermines the robustness and precision of empirical findings. The following graphs and tables illustrate the distribution of nationalities by gender category for each year.

```python
from opensanctions.const import country_mapping
from IPython.display import display
from opensanctions.h2.h2_json import json_categorical_frequency_for_sanctioned_gender_by_year

print("\033[1mTables from 1 to 6.\033[0m Relative frequency of gender by nationality per year.")
results =json_categorical_frequency_for_sanctioned_gender_by_year(category='nationality', mapping_dict=country_mapping, min_count=50)

print("\033[1mFigures from 1 to 6\033[0m\n")
for year, relative_frequency_table in results.items():
    print(f"Results for {year}:")
    display(relative_frequency_table)
```
