# Hypothesis 1: CSV processing outcomes

Hypothesis 1: *senders do not consider gender a relevant category in sanctions listing data*

The [statements.csv](https://www.opensanctions.org/docs/statements/) is a bulk file that stores information from different sources and allows us to trace the origin of each attribute of any tracked individual. For the purposes of our project, this dataset was primarily used to trace the origin of the 'gender' attribute for each sanctioned individual by distinguishing between information that came directly from official sanction lists and information collected from non-official sources (e.g. Wikidata). This was done to verify whether senders systematically include gender information about designated individuals, in order to confirm or refute our first hypothesis.

The following graphs and tables show the relevant information extracted from the dataset that led to the confirmation of hypothesis 1.

## SANCTIONED INDIVIDUALS AND GENDER INFORMATION

Only individuals labelled as 'sanctioned' were considered from the statements.csv file. These individuals can be divided into two categories: sanctioned individuals and counter-sanctioned individuals. The latter differ from the former in that they are individuals who are sanctioned by non-democratic countries, which often include pro-democracy activists, journalists and human rights proponents in their own sanctions designations. In this context, however, both categories are considered together because, regardless of the sender, these individuals appear on official sanctions lists. 

**The total number of sanctioned individuals is 40,900, of whom 9,694 are counter-sanctioned.**
